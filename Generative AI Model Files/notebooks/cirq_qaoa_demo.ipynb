```python
# Cirq QAOA Demo (MaxCut on 5-node ring)
import cirq
import numpy as np

# Build a 5-node ring graph
def ring_edges(n):
    return [(i, (i+1)%n) for i in range(n)]

n = 5
qubits = cirq.LineQubit.range(n)

# Problem unitary (cost) for MaxCut on ring
def cost_circuit(gamma):
    circuit = cirq.Circuit()
    for (u, v) in ring_edges(n):
        circuit.append(cirq.ZZ(qubits[u], qubits[v])**(gamma/np.pi))
    return circuit

# Mixer unitary
def mixer_circuit(beta):
    circuit = cirq.Circuit()
    for q in qubits:
        circuit.append(cirq.X(q)**(beta/np.pi))
    return circuit

# QAOA depth p=1
beta = 0.7
gamma = 0.8
circuit = cirq.Circuit()
# Start in |+>^n
for q in qubits:
    circuit.append(cirq.H(q))

circuit += cost_circuit(gamma)
circuit += mixer_circuit(beta)

sim = cirq.Simulator()
result = sim.run(circuit, repetitions=2000)
print(result.histogram(key=tuple(qubits)))
```
