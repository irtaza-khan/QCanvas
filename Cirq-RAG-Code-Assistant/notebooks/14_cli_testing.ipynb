# 14. CLI Testing Notebook

Demonstrates and tests the Command Line Interface (CLI) commands.

```python
import sys, os
from pathlib import Path

project_root = Path("..").resolve()
os.chdir(project_root)
print(f"Working directory: {os.getcwd()}")

cli_path = project_root / "src" / "cli" / "main.py"
outputs_dir = project_root / "notebooks" / "outputs" / "cli_output"
outputs_dir.mkdir(parents=True, exist_ok=True)
print(f"CLI: {cli_path}")
```

## Help Commands

```python
print("📋 MAIN CLI HELP")
!python {cli_path} --help
```

```python
print("\n📋 GENERATE HELP")
!python {cli_path} generate --help
```

## Generate Command

```python
print("🚀 GENERATE: Bell State Circuit")
!python {cli_path} generate "Create a 2-qubit Bell state" --algorithm teleportation --no-explain
```

## Validate Command

```python
# Create test file
test_file = outputs_dir / "test_circuit.py"
test_file.write_text("import cirq\nq = cirq.LineQubit.range(2)\nc = cirq.Circuit(cirq.H(q[0]), cirq.CNOT(q[0], q[1]))\nprint(c)")
print(f"Created: {test_file}")

print("\n✅ VALIDATE")
!python {cli_path} validate {test_file}
```

## Explain Command

```python
print("📚 EXPLAIN")
!python {cli_path} explain {test_file} --depth beginner
```
